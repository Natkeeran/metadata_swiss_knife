```python
import numpy as np
import pandas as pd
import re
```

```python
input_file = pd.read_csv('data/chelva_images.csv') 
input_file.head(2)
```

loop through all, if you see, get that row, if it start with, then it is alpha, add skip

```python
records_list = pd.DataFrame([])
new_pageset = True
for index, row in input_file.iterrows():
    item = str(row["ITEM"])
    box = row["BOX"]
    item_info = item.split(".")
    page_info = item_info[1]
    page_alpha_value = re.search('[a-zA-Z]+', page_info) 
    
    # if page is the first page of a set (indicated by a), then add, otherwise, ignore
    if page_alpha_value is not None:
        if str(page_alpha_value[0]) == "a":
            records_list = records_list.append(input_file.iloc[index])
        else:
            print("Not added: " + str(item_info))
    else:
        records_list = records_list.append(input_file.iloc[index])
```

```python
records_list.to_csv("reports/chelva_list.csv", encoding='utf-8')
```
